```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.utils import column_or_1d
```

```python
data = pd.read_csv("Hotel Reservations.csv", index_col="Booking_ID")
data.head()
```

```python
data.info()
```

```python
type_of_meal_plan_encoded = pd.get_dummies(data["type_of_meal_plan"], drop_first=True).astype(int)
data = data.join(type_of_meal_plan_encoded).drop(columns="type_of_meal_plan")
```

```python
room_type_reserved_encoded = pd.get_dummies(data["room_type_reserved"], drop_first=True).astype(int)
data = data.join(room_type_reserved_encoded).drop(columns="room_type_reserved")
```

```python
column_to_show = 'market_segment_type'
print(f"\nUnique values in the '{column_to_show}' column:")
print(data[column_to_show].unique())
```

```python
# Define the column you want to modify
column_to_modify = 'market_segment_type'

# Define the list of specific values to group into 'Other'
values_to_group = ['Corporate', 'Aviation', 'Complementary']

# Check the current value counts before grouping (optional but recommended)
print(f"Original value counts for '{column_to_modify}':")
print(data[column_to_modify].value_counts())

# Use the replace method to change the specified values to 'Other'
# Using .loc to ensure proper assignment and avoid warnings
data.loc[:, column_to_modify] = data[column_to_modify].replace(values_to_group, 'Other')

# Check the value counts after grouping to verify the change
print(f"\nValue counts for '{column_to_modify}' after grouping:")
print(data[column_to_modify].value_counts())

# Display the unique datas of the DataFrame to see the changes (optional)
column_to_show = 'market_segment_type'
print(f"\nUnique values in the '{column_to_show}' column:")
print(data[column_to_show].unique())
```

```python
market_segment_type_encoded = pd.get_dummies(data["market_segment_type"], drop_first=True).astype(int)
data = data.join(market_segment_type_encoded).drop(columns="market_segment_type")
```

```python
booking_status_encoded = pd.get_dummies(data["booking_status"], drop_first=True).astype(int)
data = data.join(booking_status_encoded).drop(columns="booking_status")
```
